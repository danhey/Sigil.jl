```julia
using OrdinaryDiffEq
init = Sigil.init_guess(1*Msun, 0.7, 0.3)
tspan = (0., star.M)


function bc2!(residual, u, p, t) # u[1] is the beginning of the time span, and u[end] is the ending
    @info(u[1])
    residual[1] = u[1][1] + 0 # solution at core for radius should be 0
    residual[2] = u[1][2] - 0 # the solution at the end of the time span should be pi/2
    residual[3] = u[end][3] - 0 # the solution at the end of the time span should be pi/2
    residual[4] = u[end][4] - 0 # the solution at the end of the time span should be pi/2
end

bvp2 = TwoPointBVProblem(Sigil.deriv!, bc2!, init, tspan)
sol2 = solve(bvp2,  Shooting(Vern7()))
plot(sol2)


# function bc3!(residual, sol, p, t)
#     @info(residual[1])
#     residual[1] = sol[1][1] + 0 # use the interpolation here, since indexing will be wrong for adaptive methods
# #     residual[2] = sol[1][1] - 0
# #     residual[3] = sol[end][1] - 0
# #     residual[4] = sol[end][1] - 0
    
# end
# bvp3 = BVProblem(Sigil.deriv!, bc3!, u₀_2, tspan)
# sol3 = solve(bvp3, Shooting(Vern7()))
```
